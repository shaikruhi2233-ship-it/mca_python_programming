```python
class Student:
    def __init__(self,name,marks=[]):
        self.name = name
        self.marks = marks
    def total_marks(self):
        return sum(self.marks)
    def __gt__(self,other):
        if isinstance(other,Student):
            return self.total_marks() > other.total_marks()
        elif isinstance(other,int):
            return self.total_marks() > other
    def __repr__(self):
        return f'Student({self.name},{self.total_marks()})'
    
# outside the function
def is_passed(marks):
    return marks > 200
students = [
    Student('Sameer',[23,56,67,20,10]),
    Student('Bahrgav',[43,86,77]),
    Student('Madhav',[83,86,87]),
    Student('Vishnu',[53,86,67]),
]    

list(filter(is_passed,students))
```
